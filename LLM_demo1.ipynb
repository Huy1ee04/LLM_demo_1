#Tạo pileline dùng để dịch ( bằng cách truyền tham số "translation" cho hàm)

```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("translation", model="google-t5/t5-base")
```

```python
text = "I am hungry, i want to eat some thing hot, such as:"
output = pipe(text)

print(output)
```

#Dùng mô hình t5_base để hỏi đáp
=> gặp lỗi do mô hình t5 chưa được huấn luyện rõ ràng cho việc hỏi đáp (QA), và cách sử dụng model qua pipeline nó bị fix cứng, không khai báo được các tham số cần thiết nên nó vẫn chưa hỏi đáp được
#Cách khắc phục: sử dụng model bằng cách tải trực tiếp thay vì dùng pipeline; đồng thời sử dụng các model khác đã được fine-tune (tinh chỉnh)

```python
pipe2 = pipeline("text2text-generation", model="google-t5/t5-base")
```

```python
question = "Question: What is the capital of France? Answer:"
answer = pipe2(question)

print(answer)
```

#Sử dụng mô hình fine-tuned từ mô hình t5-base có tên là flan-t5-base

```python
from transformers import pipeline

pipe = pipeline("text2text-generation", model="google/flan-t5-base")

question = "What is the capital of France?"
prompt = f"Answer the following question: {question}"

answer = pipe(prompt)

print(answer)
```
